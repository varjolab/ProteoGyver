```python
import os
```

```python
imps = []
for root, dirs, files in os.walk('app'):
    for f in files:
        if f.endswith('py'):
            with open(os.path.join(root,f)) as fil:
                for line in fil:
                    if line.startswith('import'):
                        imps.append([line, f])
                    elif line.startswith('from') and 'import' in line:
                        imps.append([line, f])
for i in imps:
    print(i)
```
